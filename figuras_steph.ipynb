# Practica para graficar con matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
x =np.linspace(0,6.28,100)
fig, ax = plt.subplots()
ax.plot(x,np.sin(x),c="C1",label="cos(x)")
ax.set_xlabel("pocicion infinito [mm]")
ax.set_ylabel("Funcion de onda")
ax.set_title("Funcion de onda de un pozo de potencial")
ax.legend()
ax.grid(linestyle="--")
```
